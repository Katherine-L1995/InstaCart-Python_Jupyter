### [Part 2] Task 07. Explore whether there’s a connection between age and spending power (income) using a scatterplot

```python
# Reduce previously made sample to only the columns needed
instacart_4 = small [['age','income']]
```

```python
# Create a scatterplot
scatterplot = sns.scatterplot(x = 'age', y = 'income',data = instacart_4)
```

The income distribution across different age groups reveals distinct patterns. Among individuals aged 20 to approximately 40, income levels vary widely, though there is a notable concentration at the lower end of the income spectrum ($10K-$20K). In contrast, for those aged 40 and above, there is a discernible upward trend in income levels, with a significant concentration in higher income brackets. This trend suggests a potential correlation between age and income, where income appears to increase as individuals grow older.

The scatterplot visually illustrates these trends, highlighting that the median income for individuals over 40 is considerably higher than that for those under 40. This observation implies that income may indeed rise with age. However, while the scatterplot provides a preliminary indication of this relationship, a more rigorous statistical analysis is necessary to quantify the correlation and fully understand the dynamics between age and income among the observed groups.
